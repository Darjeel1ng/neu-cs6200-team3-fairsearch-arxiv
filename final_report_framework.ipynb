## Abstract

This report investigates institutional bias in arXiv-based academic RAG systems. We examine whether baseline retrieval overrepresents privileged institutions, whether LLM synthesis amplifies this imbalance, and whether fairness-aware reranking can reduce bias while preserving utility.

## Keywords

retrieval-augmented generation, institutional bias, fairness-aware ranking, ChromaDB, academic search

## 1. Introduction

### 1.1 Motivation

Academic RAG systems retrieve papers and synthesize answers for research queries. If retrieval favors elite or high-resource institutions, underrepresented research receives lower visibility, affecting citation cycles, funding opportunities, and knowledge access.

### 1.2 Contributions

- Define institutional privilege using an external proxy (QS Top-50 CS universities).
- Build an arXiv-based retrieval pipeline for bias auditing.
- Evaluate both retrieval parity and synthesis neutrality.
- Analyze fairness-utility tradeoffs with reranking methods (MMR / fairness-aware reranking).

## 2. Background

- Problem focus: institutional bias in academic search over arXiv.
- Bias framing:
  - Privileged institutions: proxy group based on QS Top-50 CS ranking.
  - Underrepresented institutions: institutions outside this privileged list.
  - Non-university labs: treated as a separate category.
  - Missing or unclear affiliation metadata: labeled as unknown.
- Institutional homophily concern: elite visibility can create feedback loops (more citations -> higher visibility -> more citations).

## 3. Methodology

### 3.1 Data and Corpus

- Corpus: random sample of arXiv Computer Science papers.
- Document text fields: title + abstract.
- Metadata: institution, country/region, category, publication info.
- Fairness labels: privileged vs underrepresented institutions.

### 3.2 RAG Pipeline

Query -> embedding -> vector retrieval in ChromaDB -> top-k papers -> evaluation.

- Vector DB: ChromaDB (local development, Python integration).
- Embedding baseline: all-MiniLM-L6-v2.

### 3.3 Fairness Mitigation

- Baseline retrieval parity audit.
- Reranking with MMR / fairness-aware reranking.
- Compare fairness outcomes against utility metrics.

## 4. Experimental Setup

### 4.1 Research Questions

- RQ1 Retrieval Parity: does baseline retrieval overrepresent privileged institutions relative to corpus share?
- RQ2 Synthesis Neutrality: do generated answers emphasize privileged-institution papers more than underrepresented ones?
- RQ3 Fairness-Utility Tradeoff: can reranking reduce bias while preserving retrieval quality?

### 4.2 Evaluation Framework

- Compare baseline retrieval and reranked retrieval.
- Assess institutional representation in top-k results.
- Evaluate utility impact alongside fairness gains.
- Use query-level examples to inspect qualitative shifts in retrieved evidence.
